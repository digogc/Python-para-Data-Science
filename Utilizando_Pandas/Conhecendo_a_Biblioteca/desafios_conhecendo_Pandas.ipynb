## Aula 2: Análise exploratória dos dados

Para os desafios dessa aula, vamos utilizar a mesma base de dados que estamos utilizando para o desenvolvimento do nosso projeto. Sendo assim, na célula abaixo vamos realizar a importação dessa base de dados e aplicar as transformações já realizadas até o momento da aula 2.

```python
import pandas as pd
```

```python
# importando a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

# removendo imóveis comerciais
df = dados.query('@imoveis_comerciais not in Tipo')

# selecionando apenas imóveis do tipo apartamento
df = df.query('Tipo == "Apartamento"')

# visualizando a base de dados
df.head()
```

### Desafio 1
Calcular a média de quartos por apartamento

```python
# Obtendo a média de quartos por apartamento
df.Quartos.mean()
```

### Desafio 2
Conferir quantos bairros únicos existem na nossa base de dados

```python
# Visualizando quais são os bairros
df.Bairro.unique()
```

```python
# Contando a quantidade de bairros no total
df.Bairro.nunique()
```

### Desafio 3
Analisar quais bairros possuem a média de valor de aluguel mais elevadas

```python
df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)
```

Os 5 bairros mais caros são Joá, Apoador, Cidade Jardim, Ipanema e Botafogo

### Desafio 4
Criar um gráfico de barras verticais que apresente os 5 bairros com a média de valor de aluguel mais elevadas

```python
# Visualizando as distribuição da média dos aluguéis
media_valor_bairro = df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)
media_valor_bairro.plot(kind = 'barh')
```

```python
# Selecionando os 5 bairros
bairros_mais_caros = media_valor_bairro.head()
```

```python
bairros_mais_caros.plot(kind = 'bar')
```

## Aula 3

Importe novamente o arquivo [**alunos.csv**](https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv) para realizar os desafios abaixo.

```python
# Importando novamente os dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
dados.head()
```

### Desafio 1
Verifique se a base de dados possui dados nulos e caso tenha, realize o tratamento desses dados nulos

```python
# Verificando a presença de valores nulos
dados.isnull().sum()
```

```python
# Substituindo os valores nulos por zero
dados.fillna(0, inplace=True)
dados.head()
```

```python
# Verificando se os nulos foram substituidos
dados.isnull().sum()
```

### Desafio 2
Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados

```python
# Criando a variável que possui os index que devem ser removidos
ex_alunos = dados.query('Nome == "Alice" | Nome == "Carlos"').index
# Visualizando os índices, para ver se realmente foram encontrados
ex_alunos
```

```python
# Removendo os registros com o método drop
dados.drop(ex_alunos, axis=0, inplace=True)
# Realizando uma consulta para ver se os alunos foram removidos
dados.query('Nome == "Alice" | Nome == "Carlos"')
```

### Desafio 3
Aplique um filtro que selecione apenas os alunos que foram aprovados

```python
# Criando o filtro
aprovados = dados['Aprovado'] == True
# Visualizando os alunos aprovados
dados[aprovados]
```

```python
# Realizando o mesmo filtro, porém com o método query
dados.query('Aprovado == True')
```

### Desafio 4
Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv"

```python
# Criando uma variável para os alunos aprovados
df_aprovados = dados[aprovados]
```

```python
# Salvando em um csv
df_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

```python
# Lendo o csv para ver se o arquivo foi salvo corretamente
pd.read_csv('alunos_aprovados.csv')
```

### **Extra**
Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0. **Dica**: pesquise pelo método **replace**

```python
# Substuindo as notas 7 por notas 8
df_aprovados = df_aprovados.replace(7.0, 8.0)
df_aprovados
```

```python
# Salvando a alteração no arquivo csv
df_aprovados.to_csv('alunos_aprovados.csv', index=False)
# Lendo e visualizando
pd.read_csv('alunos_aprovados.csv')
```

## Aula 4

Importe novamente o arquivo [**alunos.csv**](https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv) para realizar os desafios abaixo.

```python
# Importando novamente os dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df = pd.read_csv(url)
df.head()
```

### Desafio 1
Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles

```python
df['Pontos_extras'] = df['Notas'] * 0.4
df.head()
```

### Desafio 2
Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras

```python
df['Notas_finais'] = df['Notas'] + df['Pontos_extras']
df.head()
```

```python
# Existem linhas que ficaram com valores maiores que 10, vou substituir as notas por 10
df['Notas_finais'] = df['Notas_finais'].apply(lambda x: 10 if x > 10 else x)
df.head()
```

### Desafio 3
Como os alunos ganharam pontos extras, alguns que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

* True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
* False: caso o aluno esteja reprovado (nota final deve ser menor que 6).


```python
df['Aprovado_final'] = df['Notas_finais'].apply(lambda x: True if x >= 6 else False)
df.head()
```

### Desafio 4
Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras

```python
# Utilizando o método query
df.query('Aprovado != Aprovado_final')
```

```python
# Criando uma seleção
selecao = df.Aprovado != df.Aprovado_final
df[selecao]
```
