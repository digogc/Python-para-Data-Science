# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
# o primeiro passo é relembrar como os dados estão estruturados no dataframe
dados.head()
```

As colunas que serão relevantes: 'Tipo' e 'Valor'

```python
# aplicando o método pd.mean() para todos os valores
dados['Valor'].mean()
```

```python
# Agrupando os dados por 'Tipo' e extraindo a média por tipo de alguel
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
# Selecionando apenas a coluna das médias de Valor
dados.groupby('Tipo')['Valor'].mean()
```

O que é retornado nesse caso é uma Series, mas é possível alterar isso

```python
# Transformando o resultado em um dataframe e ordenando os valores
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Para visualizar melhor as informações, podemos plotar um gráfico
# Salvando em uma variável
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

# Plotando o dataframe
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```

Existem imóveis residenciais e comerciais, resultando em uma grande variação na média dos valore \\
Essa variação pode ser prejudicial para o modelo que será gerado para gerar o preço dos imóveis. Portanto, é uma boa ideia utilizar apenas um tipo de imóvel (no caso, o residencial)

# Manipulando os dados

É necessário importar novamente a base de dados, porque vamos começar uma demanda para outro time, e para isso, é necessário ter todos os dados, e não somente os dados referentes aos apartamentos.

## Criando colunas numéricas

```python
# Salvando a url dos dados em uma variável
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
# Salvando os dados em uma variável
dados = pd.read_csv(url, sep=';')
# Conferindo para ver se os dados foram baixados e lidos corretamente
dados.head()
```

```python
# Criando uma coluna para valor por mês, somando o aluguel com o condomínio
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
# Visualizando os dados
dados.head()
```

```python
# Criando a coluna de valor por ano, que é valor por mês * 12 + IPTU
dados['Valor_por_ano'] = dados['Valor_por_mes']*12 + dados['IPTU']
# Visualizando os dados
dados.head()
```

## Criando colunas categóricas

```python
# Criando uma nova coluna com uma descrição
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
# Adicionando todos os valores solicitados na descrição
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'

# Visualizando o dataframe
dados.head()
```

```python
# Criando e populando a coluna que indica se há suítes ou não
dados['Possui_suite'] = dados['Suites'].apply(lambda x : 'Sim' if x > 0 else 'Não')
# Visualizando o novo dataframe
dados.head()
```

```python
# Salvando o df em um arquivo
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
# Vendo se foi salvo corretamente
pd.read_csv('dados_completos_dev.csv', sep=';')
```
