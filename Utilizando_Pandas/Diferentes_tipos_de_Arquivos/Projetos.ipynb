# Trabalhando com Banco de Dados

```python
import sqlalchemy
```

```python
# importante métodos da biblioteca
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Criando o banco de dados
engine = create_engine('sqlite:///:memory:')
```

```python
# Obtendo os dados
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'

dados = pd.read_csv(url)
```

```python
# Visualizando os dados
dados.head()
```

```python
# Salvando como um banco de dados
dados.to_sql('clientes', engine, index=False)
```

```python
# Criando um inspector para visualizar o banco de dados
inspector = inspect(engine)
```

```python
# Vendo quais são as tabelas criadas no banco de dados
print(inspector.get_table_names())
```

```python
# Criando a consulta sql
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado" '
```

```python
# Lendo a consulta sql
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
# Salvando a nossa consulta como uma nova tabela
empregados.to_sql('empregados', con=engine, index=False)
```

```python
# Lendo a tabela inteira
pd.read_sql_table('empregados', engine)
```

```python
# Lendo apenas algumas colunas da tabela
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
# importanto o método text
from sqlalchemy import text
```

```python
# criando uma consulta para visualizar toda a tabela de clientes
query = 'SELECT * FROM clientes'

# Realizando a consulta
pd.read_sql(query, engine)
```

```python
# Removendo um registro do banco de dados
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
# Abrindo uma conexão com o banco de dados para realizar a consulta
with engine.connect() as conn:
  # executando a consulta
  result = conn.execute(text(query))
  # confirmando a consulta
  conn.commit()
```

```python
# Visualizando a tabela para ver se o cliente foi realmente excluído
pd.read_sql_table('clientes', engine)
```

```python
# Atualizando dados do cliente 5008808
query = 'UPDATE clientes SET Grau_escolaridade = "Ensino Superior" WHERE ID_Cliente= 5008808'

# Abrindo uma conexão com o banco de dados para realizar a consulta
with engine.connect() as conn:
  # executando a consulta
  result = conn.execute(text(query))
  # confirmando a consulta
  conn.commit()
```
