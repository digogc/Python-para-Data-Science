# Desafio criando um banco de dados

```python
# importando a biblioteca sqlalchemy e os métodos que serão necessários
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Salvando o salvando o arquivo csv como um df
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados = pd.read_csv(url)

dados.head()
```

```python
# Criando um banco de dados
engine = create_engine('sqlite:///:memory:')
```

```python
# Movendo os dados do df para o banco de dados
dados.to_sql('clientes', engine, index=False)
```

```python
# criando um inspector para conseguir obter dados do bd
inspector = inspect(engine)

# Verificando se a tabela foi criada corretamente
print(inspector.get_columns('clientes'))
```

```python
# Visualizando a tabela criada
query = 'SELECT * FROM clientes'

pd.read_sql(query, engine)
```

## Realizando alterações no banco de dados:
1a) Alteração: \
Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.

```python
# importando o método text
from sqlalchemy import text
```

```python
# Criando a consulta que irá realizar a atualização
query = 'UPDATE clientes SET Rendimento_anual = 300000 WHERE ID_Cliente = 6840104'

# Abrindo uma conexão com o banco de dados para realizar a consulta
with engine.connect() as conn:
  # executando a consulta
  result = conn.execute(text(query))
  # confirmando a consulta
  conn.commit()
```

```python
# Visualizando a linha alterada
query = 'SELECT * FROM clientes WHERE ID_Cliente = 6840104'

pd.read_sql(query, engine)
```
